# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python
# read my data in last part
df=pd.read_csv('CityBike_Station_Riga.csv')
```

```python
df
```

```python
# load my key in env
MY_FQ_API_KEY = os.getenv('MY_FQ_API_KEY')
```

```python
# set the headers for data get process
headers = {
    "accept": "application/json",
    "Authorization": MY_FQ_API_KEY}
```

```python
# load lat&lon info, loop for get building info within a 1000m radius.
for index, row in df.iterrows():
    station_name = row['Station Name']
    latitude = row['Latitude']
    longitude = row['Longitude']

    response = requests.get(
        f'https://api.foursquare.com/v3/places/search?ll={latitude},{longitude}&radius=1000', headers=headers
    )

    
    all_data = response.json()
```

```python
all_data
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
for index, row in df.iterrows():
    station_name = row['Station Name']
    latitude = row['Latitude']
    longitude = row['Longitude']
# add query=bars
    response = requests.get(
        f'https://api.foursquare.com/v3/places/search?query=bars&ll={latitude},{longitude}&radius=1000', headers=headers
    )

    
    data = response.json()
data
```

```python
places = data.get('results', [])
bars = []
for place in places:
    categories = place.get('categories', [])
    for category in categories:
        bar_info = {
            'name': place.get('name'),
            'location': place.get('location', {}).get('formatted_address'),
            'latitude': place.get('geocodes', {}).get('main', {}).get('latitude'),
            'longitude': place.get('geocodes', {}).get('main', {}).get('longitude'),
            'distance': place.get('distance'),
        }

        bars.append(bar_info)
```

```python
bars
```

Put your parsed results into a DataFrame

```python
FQ_df = pd.DataFrame(columns=['Station Name', 'Bars'])

for index, row in df.iterrows():
    station_name = row['Station Name']
    latitude = row['Latitude']
    longitude = row['Longitude']
    # add query=bars
    response = requests.get(
        f'https://api.foursquare.com/v3/places/search?query=bars&ll={latitude},{longitude}&radius=1000', headers=headers
    )

    
    data = response.json()
    places = data.get('results', [])
    bars = []
    for place in places:
        categories = place.get('categories', [])
        for category in categories:
            bar_info = {
                'name': place.get('name'),
                'location': place.get('location', {}).get('formatted_address'),
                'latitude': place.get('geocodes', {}).get('main', {}).get('latitude'),
                'longitude': place.get('geocodes', {}).get('main', {}).get('longitude'),
                'distance': place.get('distance'),
            }

            bars.append(bar_info)
    new_row = pd.DataFrame({'Station Name': [station_name], 'Bars': [bars]})
    FQ_df = pd.concat([FQ_df, new_row], ignore_index=True)

FQ_df                
```

```python
FQ_df = FQ_df.explode('Bars')

bars_info = FQ_df['Bars'].apply(pd.Series)

FQ_df = pd.concat([FQ_df.drop(['Bars'], axis=1), bars_info], axis=1)

FQ_df
```

```python
FQ_df = FQ_df.rename(columns={
    'name': 'Bars_name',
    'location': 'Bars_location',
    'latitude': 'Bars_latitude',
    'longitude': 'Bars_longitude'
})


FQ_df
```

```python
FQ_df.to_csv('FQ_df.csv', index=False)
```
